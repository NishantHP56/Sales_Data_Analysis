```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
sales = pd.read_csv('/content/sales_data.csv')
sales
```

```python
sales_data=sales
sales_data
```

```python
data_sample = sales
data_sample
```

```python
sales['Sale_Date'] = pd.to_datetime(sales['Sale_Date'])
daily_sales = sales.groupby('Sale_Date')['Sales_Amount'].sum().reset_index()
daily_sales.head()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=daily_sales, x='Sale_Date', y='Sales_Amount')
plt.title('Daily Total Sales Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

category_sales = sales.groupby('Product_Category')['Sales_Amount'].sum().reset_index()
best_selling_category = category_sales.sort_values(by='Sales_Amount', ascending=False)

print("Best-selling Product Categories by Total Sales Amount:")
display(best_selling_category.head())
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Sales_Amount', y='Product_Category', data=best_selling_category, palette='viridis')
plt.title('Total Sales Amount by Product Category')
plt.xlabel('Total Sales Amount')
plt.ylabel('Product Category')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
